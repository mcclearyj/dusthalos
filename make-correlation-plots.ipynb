## IF YOU ONLY TAKE TWO REDSHIFT BINS, GET CLOSER TO LITERATURE RESULT

```python
dd=fitsio.read('/Users/jemcclea/Research/DES_DUST/round2/dusthalos/outputs/dust_correlation_dd2.fits') # The dust signal -- real fg x real background 
dr=fitsio.read('/Users/jemcclea/Research/DES_DUST/round2/dusthalos/outputs/dust_correlation_dr2.fits') # fg x random background
fr=fitsio.read('/Users/jemcclea/Research/DES_DUST/round2/dusthalos/outputs/dust_correlation_fr2.fits') # random fg x background
rr=fitsio.read('/Users/jemcclea/Research/DES_DUST/round2/dusthalos/outputs/dust_correlation_rr2.fits') # random fg x random background
```

```python
fig = plt.figure(figsize=(14,7))

### in log space
ax=fig.add_subplot(121)
ax.errorbar(dd['meanR'],dd['kappa']-(dr['kappa']-np.mean(dr['kappa']))-fr['kappa'],yerr=dd['sigma'],label='random subtracted 2 z bins')
ax.errorbar(dd['meanR'],dd['kappa'],yerr=dd['sigma'],label='raw 2 z bins')
ax.set_xscale('log')
ax.set_yscale('log')

ax.set_xlim(.1,250)
r = np.logspace(-2,2.7,10)
av2 = 2.4e-3 * (r/3.06)**(-0.84)
ax.plot(r,av2,label='Scaled Menard (2010)')

#Uncomment to get axis in physical scale: kpc
#a=ax.get_xticks().tolist()
#print a
#physical=[a[i]*0.0279 for i in range(len(a))]
#ax.set_xticklabels(physical)

ax.set_xlabel('target separation (kpc)')
ax.set_ylabel('A_v (mag)')
ax.legend()

### in linear space
ax2=fig.add_subplot(122)
ax2.errorbar(dd['meanR'],dd['kappa']-(dr['kappa']-np.mean(dr['kappa'])),yerr=dd['sigma'],label='random subtracted 2 z bins')
ax2.errorbar(dd['meanR'],dd['kappa'],yerr=dd['sigma'],label='raw 2 z bins')

ax2.set_xlim(0.1,200)
#ax2.set_ylim(-7e-4,0.)
ax2.plot(r,av2,label='scaled Menard (2010)')
ax2.set_xscale('log')
ax2.axhline(0,color='black',linestyle='--',alpha=0.5)
ax2.set_xlabel('target separation (arcmin)')
ax2.legend()
fig.savefig('/Users/jemcclea/Research/DES_DUST/round2/dusthalos/outputs/correlationFuncFigures/dust_measurement_2zbin.png')
```
