# How self-consistent is the A_v of the foreground wiseScosSvm sample?

```python
cat=Table.read('../../catalogs/Sscom_exactArea_galzCut.fits',format='fits')
```

```python
data=cat
B = data['bCalCorr']
R = data['rCalCorr']
W1 = data['w1mCorr']
W2 = data['w2mCorr']
data = np.vstack([B,R,W1,W2])
covar = np.cov(data)


dmdp = np.array([1.52925446, 0.81354838, 0.03857676, 0.02252133])

delta = (np.zeros_like(data).T + dmdp)
Cinv = np.linalg.inv(covar)
colors = (data.T - np.average(data,axis=1)).T
est = np.sum(delta.T*np.dot(Cinv,colors),axis=0)/(np.dot(dmdp,np.dot(Cinv,dmdp)))
wt = 1./(np.dot(dmdp,np.dot(Cinv,dmdp)))
```

```python
est
```

```python
print (wt)

step1=np.dot(Cinv,dmdp)
np.shape(step1)
var_rho=np.dot(dmdp.T,step1)
print(var_rho)
```

```python
## OK, these are fairly tight, but not remarkably so.
fig=plt.figure(figsize=(8,7))
ax=fig.add_subplot(111)
_=ax.hist(est,bins=100,range=[-1.5,1.5])
ax.set_xlabel('Av of wiseScosSvm foreground sample')
ax.annotate('variance ~ 10%',[.75,.85],xycoords='figure fraction')
ax.axvline(0,color='black',linestyle='--',alpha=0.5)
fig.savefig('correlationFuncFigures/av_wiseScosSvm.png')
```
